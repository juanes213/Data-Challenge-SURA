```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Conversión de fechas
df['FECHA_ATENCION'] = pd.to_datetime(df['FECHA_ATENCION'], dayfirst=True, errors='coerce')
df['FECHA_PROCESO'] = pd.to_datetime(df['FECHA_PROCESO'], dayfirst=True, errors='coerce')

# Extracción de año y mes
df['Year_Atencion'] = df['FECHA_ATENCION'].dt.year
df['Month_Atencion'] = df['FECHA_ATENCION'].dt.month
df['Year_Proceso'] = df['FECHA_PROCESO'].dt.year
df['Month_Proceso'] = df['FECHA_PROCESO'].dt.month

# Filtro de años
df_filtrado = df[
    (df['Year_Atencion'].between(2019, 2024)) &
    (df['Year_Proceso'].between(2019, 2024))
]

# Agrupación
atencion_counts = df_filtrado.groupby(['Year_Atencion', 'Month_Atencion']).size().reset_index(name='Counts')
proceso_counts = df_filtrado.groupby(['Year_Proceso', 'Month_Proceso']).size().reset_index(name='Counts')

# Guardar en JSON
atencion_counts.to_json('atencion_counts.json', orient='records', indent=4)
proceso_counts.to_json('proceso_counts.json', orient='records', indent=4)

# Graficar
plt.figure(figsize=(14, 6))

# Gráfico de atención médica
plt.subplot(1, 2, 1)
sns.lineplot(
    data=atencion_counts,
    x='Month_Atencion',
    y='Counts',
    hue='Year_Atencion',
    marker='o',
    palette='Set1'
)
plt.title('Monthly Medical Attention (2019–2024)', fontsize=20)
plt.xlabel('Month', fontsize=16)
plt.ylabel('Number of Attendances', fontsize=16)
plt.xticks(range(1, 13))

# Gráfico de procesos
plt.subplot(1, 2, 2)
sns.lineplot(
    data=proceso_counts,
    x='Month_Proceso',
    y='Counts',
    hue='Year_Proceso',
    marker='o',
    palette='Set1'
)
plt.title('Monthly Processes (2019–2024)', fontsize=20)
plt.xlabel('Month', fontsize=16)
plt.ylabel('Number of Processes', fontsize=16)
plt.xticks(range(1, 13))

plt.tight_layout()
plt.show()
```

```python
# Handle missing values in incapacity days columns
incapacity_cols = ['Mean_Incapacity_Days', 'Median_Incapacity_Days', 'Total_Incapacity_Days']
for col in incapacity_cols:
    # Fill missing values with the median of the column grouped by Service_Type
    median_by_service = df_processed.groupby('Service_Type')[col].transform('median')
    df_processed[col] = df_processed[col].fillna(median_by_service)
    # If there are still NaNs (for service types with all NaN), fill with 0
    df_processed[col] = df_processed[col].fillna(0)

# Create additional time-based features
df_processed['Year_Month'] = df_processed['Date'].dt.strftime('%Y-%m')
df_processed['Quarter'] = df_processed['Date'].dt.quarter
df_processed['Month_of_Year'] = df_processed['Date'].dt.month
df_processed['Year_Fraction'] = df_processed['Year'] + (df_processed['Month'] - 1) / 12

# Create seasonality features using sine and cosine transformations
# These capture the cyclical nature of months and quarters
df_processed['Month_sin'] = np.sin(2 * np.pi * df_processed['Month_of_Year'] / 12)
df_processed['Month_cos'] = np.cos(2 * np.pi * df_processed['Month_of_Year'] / 12)
df_processed['Quarter_sin'] = np.sin(2 * np.pi * df_processed['Quarter'] / 4)
df_processed['Quarter_cos'] = np.cos(2 * np.pi * df_processed['Quarter'] / 4)
```

```python
label_encoders = {}
categorical_cols = ['Municipality', 'Service_Type']

for col in categorical_cols:
    le = LabelEncoder()
    df_processed[f'{col}_encoded'] = le.fit_transform(df_processed[col])
    label_encoders[col] = le
    # Keep original columns for interpretability

# Create lagged features for time series modeling
# Group by municipality and service type to create proper lags
for lag in [1, 2, 3, 6, 12]:  # 1, 2, 3, 6 months and 1 year lags
    # Sort data chronologically within each group
    df_processed = df_processed.sort_values(['Municipality', 'Service_Type', 'Date'])
    
    # Create lag features for each group
    df_processed[f'Service_Count_lag_{lag}'] = df_processed.groupby(['Municipality', 'Service_Type'])['Service_Count'].shift(lag)
```

```python
# Create rolling statistics as features
# First sort the data chronologically
df_processed = df_processed.sort_values(['Municipality', 'Service_Type', 'Date'])

# Then compute rolling statistics for each group
for window in [3, 6, 12]:
    df_processed[f'Service_Count_rolling_mean_{window}'] = df_processed.groupby(['Municipality', 'Service_Type'])['Service_Count'].transform(
        lambda x: x.rolling(window=window, min_periods=1).mean()
    )
    df_processed[f'Service_Count_rolling_std_{window}'] = df_processed.groupby(['Municipality', 'Service_Type'])['Service_Count'].transform(
        lambda x: x.rolling(window=window, min_periods=1).std()
    )

# Calculate growth rates
df_processed['Growth_Rate_MoM'] = df_processed.groupby(['Municipality', 'Service_Type'])['Service_Count'].pct_change()
df_processed['Growth_Rate_YoY'] = df_processed.groupby(['Municipality', 'Service_Type'])['Service_Count'].pct_change(12)

# Replace infinities and NaNs in growth rate columns
growth_cols = ['Growth_Rate_MoM', 'Growth_Rate_YoY']
for col in growth_cols:
    df_processed[col] = df_processed[col].replace([np.inf, -np.inf], np.nan)
    df_processed[col] = df_processed[col].fillna(0)

# Calculate days since first service for each municipality-service combination
df_processed['First_Service_Ever'] = df_processed.groupby(['Municipality', 'Service_Type'])['First_Service_Date'].transform('min')
df_processed['Days_Since_First_Service'] = (df_processed['Date'] - df_processed['First_Service_Ever']).dt.days

# Calculate consecutive months with services for each municipality-service combination
df_processed['Month_Sequence'] = df_processed.groupby(['Municipality', 'Service_Type']).cumcount() + 1
```

```python
# Create features representing the time dimension and cyclical nature of data
# This can help capture long-term trends and seasonality
current_date = pd.Timestamp.now()
df_processed['Days_From_Now'] = (df_processed['Date'] - current_date).dt.days
df_processed['Is_Holiday_Month'] = df_processed['Month'].isin([1, 7, 12]).astype(int)  # Common holiday months

# Fill NaNs in lag features
lag_cols = [col for col in df_processed.columns if 'lag' in col or 'rolling' in col]
for col in lag_cols:
    df_processed[col] = df_processed[col].fillna(0)

# Create a feature for anomaly detection
df_processed['Is_Anomaly'] = 0
for (muni, service), group in df_processed.groupby(['Municipality', 'Service_Type']):
    if len(group) >= 12:  # Need enough data points
        mean = group['Service_Count'].mean()
        std = group['Service_Count'].std()
        if std > 0:  # Avoid division by zero
            z_scores = (group['Service_Count'] - mean) / std
            df_processed.loc[group.index, 'Is_Anomaly'] = (abs(z_scores) > 3).astype(int)  # Mark as anomaly if z-score > 3
```

### 1. ARIMA MODEL (for a selected municipality and service type combination)

```python
# Select a popular municipality and service type for demonstration
top_municipality = train_data['Municipality'].value_counts().index[0]
top_service = train_data['Service_Type'].value_counts().index[0]

# Filter data for the selected municipality and service type
arima_data = train_data[(train_data['Municipality'] == top_municipality) & 
                        (train_data['Service_Type'] == top_service)]
arima_data = arima_data.sort_values('Date')

# Set up time series data
arima_ts = arima_data.set_index('Date')['Service_Count']
print(f"ARIMA training on {top_municipality} - {top_service} with {len(arima_ts)} observations")

# Determine maximum parameter orders based on data size
ts_length = len(arima_ts)
print(f"Time series length: {ts_length}")

# For SARIMA with seasonal period s, we need at least 2*(p+d+q)+(P+D+Q)*s observations
# Set limits based on available data to avoid ValueError
max_nonseasonal_param = max(min(ts_length // 4, 2), 0)  # Conservative limit
max_seasonal_param = max(min(ts_length // 24, 1), 0)    # Even more conservative for seasonal
seasonal_period = min(12, ts_length // 4)               # Adjust seasonal period if necessary

print(f"Using max non-seasonal params (p,d,q): {max_nonseasonal_param}")
print(f"Using max seasonal params (P,D,Q): {max_seasonal_param}")
print(f"Using seasonal period: {seasonal_period}")

# Check for stationarity using Augmented Dickey-Fuller test
from statsmodels.tsa.stattools import adfuller

def check_stationarity(timeseries):
    if len(timeseries) < 10:  # Need sufficient data points
        print("Time series too short for stationarity test")
        return False
        
    result = adfuller(timeseries.dropna())
    print(f'ADF Statistic: {result[0]:.4f}')
    print(f'p-value: {result[1]:.4f}')
    print('Critical Values:')
    for key, value in result[4].items():
        print(f'\t{key}: {value:.4f}')
    # If p-value > 0.05, data is non-stationary
    if result[1] > 0.05:
        print("Time series is non-stationary (fail to reject H0)")
    else:
        print("Time series is stationary (reject H0)")
    return result[1] <= 0.05

is_stationary = check_stationarity(arima_ts)

# If we have very limited data, consider simpler models
if ts_length < 30:
    print("WARNING: Limited data available. Using simple model instead of SARIMA.")
    # Fit simpler model (e.g., simple ARIMA(1,1,1) without seasonality)
    try:
        simple_model = SARIMAX(
            arima_ts,
            order=(1, 1, 1),
            seasonal_order=(0, 0, 0, 0),  # No seasonality
            enforce_stationarity=False,
            enforce_invertibility=False
        )
        simple_results = simple_model.fit(disp=False)
        
        # Use this model for forecasts
        arima_validation = valid_data[(valid_data['Municipality'] == top_municipality) & 
                             (valid_data['Service_Type'] == top_service)]
        arima_validation = arima_validation.sort_values('Date')
        
        if not arima_validation.empty:
            forecast_periods = len(arima_validation)
            arima_forecast = simple_results.get_forecast(steps=forecast_periods)
            arima_pred = arima_forecast.predicted_mean
            
            # Evaluate
            arima_actual = arima_validation.set_index('Date')['Service_Count']
            arima_mae = mean_absolute_error(arima_actual, arima_pred)
            arima_rmse = np.sqrt(mean_squared_error(arima_actual, arima_pred))
            
            print(f"Simple ARIMA Model MAE: {arima_mae:.2f}")
            print(f"Simple ARIMA Model RMSE: {arima_rmse:.2f}")
            
            # Plot
            plt.figure(figsize=(12, 6))
            plt.plot(arima_actual.index, arima_actual, label='Actual')
            plt.plot(arima_actual.index, arima_pred, label='Simple ARIMA Forecast')
            plt.title(f'Simple ARIMA Forecast vs Actual for {top_municipality} - {top_service}')
            plt.legend()
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.savefig('simple_arima_forecast.png')
            plt.close()
    except Exception as e:
        print(f"Error with simple model: {e}")
else:
    # For more data, use grid search with adjusted parameters
    import itertools
    import warnings
    warnings.filterwarnings("ignore")  # Suppress convergence warnings
    
    def grid_search_arima_params(ts, max_p=1, max_d=1, max_q=1, max_P=1, max_D=1, max_Q=1, s=12):
        best_aic = float('inf')
        best_params = None
        best_model = None
        
        # Si no es estacionaria, usar al menos d=1
        min_d = 0 if is_stationary else 1
        min_d = min(min_d, max_d)  # Asegurar que min_d no exceda max_d
        
        # Parámetros no estacionales - rangos pequeños para búsqueda exhaustiva
        p_range = range(0, max_p + 1)
        d_range = range(min_d, max_d + 1)
        q_range = range(0, max_q + 1)
        
        # Parámetros estacionales - también rangos pequeños
        P_range = range(0, max_P + 1)
        D_range = range(0, max_D + 1)
        Q_range = range(0, max_Q + 1)
        
        # Crear todas las combinaciones posibles
        all_combinations = list(itertools.product(p_range, d_range, q_range, P_range, D_range, Q_range))
        total_combinations = len(all_combinations)
        print(f"Evaluando exhaustivamente {total_combinations} combinaciones SARIMA...")
        
        # No hacemos muestreo aleatorio - probamos todas las combinaciones
        for params in all_combinations:
            p, d, q, P, D, Q = params
            
            # Calcular mínimo de observaciones requeridas para esta combinación
            min_obs_required = 2 * (p + d + q) + (P + D + Q) * s
            if len(ts) <= min_obs_required + 5:  # Añadir margen para evitar error
                continue
                
            try:
                model = SARIMAX(
                    ts,
                    order=(p, d, q),
                    seasonal_order=(P, D, Q, s) if s > 0 else (0, 0, 0, 0),
                    enforce_stationarity=False,
                    enforce_invertibility=False
                )
                results = model.fit(disp=False, maxiter=50)
                aic = results.aic
                
                if aic < best_aic:
                    best_aic = aic
                    best_params = (p, d, q, P, D, Q, s)
                    best_model = results
                    print(f"Nuevo mejor SARIMA({p},{d},{q})({P},{D},{Q},{s}) AIC={aic:.2f}")
            except Exception as e:
                print(f"Error con SARIMA({p},{d},{q})({P},{D},{Q},{s}): {e}")
                continue
        
        return best_params, best_model, best_aic
    
    # Búsqueda exhaustiva con un rango ligeramente mayor para parámetros clave
    best_params, best_model, best_aic = grid_search_arima_params(
        arima_ts, 
        max_p=2,          # Permitir hasta AR(2)
        max_d=1,          # Mantener diferenciación simple (0 o 1)
        max_q=2,          # Permitir hasta MA(2)
        max_P=1,          # Mantener componente estacional simple
        max_D=1,          # Mantener diferenciación estacional simple
        max_Q=1,          # Mantener componente estacional simple
        s=12              # Mantener período estacional de 12
    )
    
    if best_model is not None:
        print(f"\nBest SARIMA model: SARIMA{best_params} with AIC={best_aic:.2f}")
        
        # Proceed with validation
        arima_validation = valid_data[(valid_data['Municipality'] == top_municipality) & 
                         (valid_data['Service_Type'] == top_service)]
        arima_validation = arima_validation.sort_values('Date')
        
        if not arima_validation.empty:
            # Create a date range for prediction
            forecast_periods = len(arima_validation)
            arima_forecast = best_model.get_forecast(steps=forecast_periods)
            arima_pred = arima_forecast.predicted_mean
            
            # Evaluate the ARIMA model
            arima_actual = arima_validation.set_index('Date')['Service_Count']
            arima_mae = mean_absolute_error(arima_actual, arima_pred)
            arima_rmse = np.sqrt(mean_squared_error(arima_actual, arima_pred))
            
            print(f"ARIMA Model MAE: {arima_mae:.2f}")
            print(f"ARIMA Model RMSE: {arima_rmse:.2f}")
            
            # Plot results
            plt.figure(figsize=(12, 6))
            plt.plot(arima_actual.index, arima_actual, label='Actual')
            plt.plot(arima_actual.index, arima_pred, label='ARIMA Forecast')
            plt.title(f'ARIMA Forecast vs Actual for {top_municipality} - {top_service}')
            plt.legend()
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.savefig('arima_forecast_vs_actual.png')
            plt.close()
    else:
        print("Could not find suitable ARIMA parameters. Try a simpler model.")
```

### 2. XGBoost MODEL (for all municipality-service combinations)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import RandomizedSearchCV, TimeSeriesSplit
import xgboost as xgb
import seaborn as sns

# Paso 0: Volatilidad
std_df = train_data.groupby(['Municipality','Service_Type'])['Service_Count'].std().reset_index(name='std')
q1, q3 = std_df['std'].quantile([0.25,0.75])
std_df['vol_group'] = pd.cut(std_df['std'], bins=[-np.inf, q1, q3, np.inf], labels=['low','medium','high'])
train_data = train_data.drop(columns=['vol_group'], errors='ignore')
valid_data = valid_data.drop(columns=['vol_group'], errors='ignore')
train_data = train_data.merge(std_df[['Municipality','Service_Type','vol_group']], on=['Municipality','Service_Type'])
valid_data = valid_data.merge(std_df[['Municipality','Service_Type','vol_group']], on=['Municipality','Service_Type'])

# Paso 1: Features
def prepare_features(data, drop_short_lags=True):
    cols = [
        'Year','Month','Quarter','Month_sin','Month_cos',
        'Municipality_encoded','Service_Type_encoded',
        'Service_Count_lag_3','Service_Count_lag_6','Service_Count_lag_12',
        'Service_Count_rolling_mean_3','Service_Count_rolling_std_3',
        'Service_Count_rolling_mean_6','Service_Count_rolling_std_6',
        'Service_Count_rolling_mean_12','Service_Count_rolling_std_12',
        'Growth_Rate_MoM','Growth_Rate_YoY',
        'Days_Since_First_Service','Month_Sequence','Is_Holiday_Month'
    ]
    if not drop_short_lags:
        cols = ['Service_Count_lag_1','Service_Count_lag_2'] + cols
    X = data[cols].copy()
    y = data['Service_Count'].copy()
    return X, y

X_train, y_train = prepare_features(train_data)
X_valid, y_valid = prepare_features(valid_data)
scaler = StandardScaler()
num_cols = X_train.select_dtypes(include=['float64','int64']).columns
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_valid[num_cols] = scaler.transform(X_valid[num_cols])

# Paso 2: RandomizedSearchCV
param_dist = {
    'n_estimators': [100,300,500,800,1000],
    'max_depth': [3,4,5,6,8],
    'learning_rate': [0.01,0.03,0.05,0.1],
    'subsample': [0.6,0.7,0.8,0.9,1.0],
    'colsample_bytree': [0.6,0.7,0.8,0.9,1.0],
    'gamma': [0,0.1,0.5,1,5],
    'min_child_weight': [1,3,5,7],
    'reg_alpha': [0,0.1,1,5],
    'reg_lambda': [0.5,1,2,5]
}
base = xgb.XGBRegressor(objective='reg:squarederror', random_state=42, verbosity=0)
search = RandomizedSearchCV(base, param_dist, n_iter=50, scoring='neg_mean_squared_error', cv=5, n_jobs=-1, verbose=1, random_state=42)
print("Ejecutando RandomizedSearchCV...")
search.fit(X_train, y_train)
best_model = search.best_estimator_
print(f"Mejores parámetros: {search.best_params_}")

# Paso 3: Expanding window
ts = TimeSeriesSplit(n_splits=5)
rmse_scores = []
for train_idx, test_idx in ts.split(X_train):
    mdl = xgb.XGBRegressor(**search.best_params_, objective='reg:squarederror', random_state=42)
    mdl.fit(X_train.iloc[train_idx], y_train.iloc[train_idx])
    preds = mdl.predict(X_train.iloc[test_idx])
    rmse_scores.append(np.sqrt(mean_squared_error(y_train.iloc[test_idx], preds)))
print("Expanding Window RMSE por fold:", rmse_scores)
print("Promedio RMSE Expanding:", np.mean(rmse_scores))

# Paso 4: Entrenamiento final
best_model.fit(X_train, y_train)
y_train_pred = best_model.predict(X_train)
y_valid_pred = best_model.predict(X_valid)
valid_data['XGB_Pred'] = y_valid_pred

def mk_metrics(name, y_t, y_p):
    return {
        'Dataset': name,
        'MAE': mean_absolute_error(y_t, y_p),
        'RMSE': np.sqrt(mean_squared_error(y_t, y_p)),
        'R2': r2_score(y_t, y_p)
    }
stats_df = pd.DataFrame([
    mk_metrics('Train', y_train, y_train_pred),
    mk_metrics('Valid', y_valid, y_valid_pred)
])
print("\nMétricas globales:")
print(stats_df)

# Paso 5: Métricas por serie
series_std = train_data.groupby(['Municipality','Service_Type'])['Service_Count'].std().reset_index(name='std')
volatile = series_std[series_std['std'] > 0][['Municipality','Service_Type']]
metrics = []
for (muni, svc), grp in valid_data.groupby(['Municipality','Service_Type']):
    if (muni, svc) not in set(map(tuple, volatile.values)):
        continue
    yt, yp = grp['Service_Count'], grp['XGB_Pred']
    metrics.append({
        'Municipality': muni,
        'Service_Type': svc,
        'MAE': mean_absolute_error(yt, yp),
        'RMSE': np.sqrt(mean_squared_error(yt, yp)),
        'R2': r2_score(yt, yp)
    })
met_df = pd.DataFrame(metrics)
print("\nMétricas promedio series no constantes:")
print(met_df[['MAE','RMSE','R2']].mean())

# Paso 6: Por grupo de volatilidad
group_results = []
for vol in ['low','medium','high']:
    sub_tr = train_data[train_data['vol_group']==vol]
    sub_va = valid_data[valid_data['vol_group']==vol]
    if sub_tr.empty:
        continue
    X_t, y_t = prepare_features(sub_tr)
    X_v, y_v = prepare_features(sub_va)
    sc_g = StandardScaler()
    cols_g = X_t.select_dtypes(include=['float64','int64']).columns
    X_t[cols_g] = sc_g.fit_transform(X_t[cols_g])
    X_v[cols_g] = sc_g.transform(X_v[cols_g])
    m = xgb.XGBRegressor(**search.best_params_, objective='reg:squarederror', random_state=42)
    m.fit(X_t, y_t)
    p_t = m.predict(X_t)
    p_v = m.predict(X_v)
    group_results.append({
        'Volatility': vol,
        'Train_RMSE': np.sqrt(mean_squared_error(y_t, p_t)),
        'Valid_RMSE': np.sqrt(mean_squared_error(y_v, p_v)),
        'Train_R2': r2_score(y_t, p_t),
        'Valid_R2': r2_score(y_v, p_v)
    })
grp_df = pd.DataFrame(group_results)
print("\nResultados por grupo de volatilidad:")
print(grp_df)

# Paso 7: Feature importance
plt.figure(figsize=(12,8))
xgb.plot_importance(best_model, max_num_features=20, height=0.8)
plt.title('Feature Importance (Global)')
plt.tight_layout()
plt.show()

# Paso 8: Scatter RMSE vs R² con etiquetas a outliers
plt.figure(figsize=(14, 8))
sns.scatterplot(data=met_df, x='RMSE', y='R2', alpha=0.6)
plt.title('Desempeño por Serie (Municipality - Service_Type)')
plt.xlabel('RMSE')
plt.ylabel('R²')

# Etiquetas en outliers
for _, row in met_df.iterrows():
    if row['RMSE'] > 10 or row['R2'] < 0.5:
        label = f"{row['Municipality']} - {row['Service_Type']}"
        plt.text(row['RMSE'] + 0.3, row['R2'] + 0.01, label, fontsize=8)

plt.tight_layout()
plt.show()
```

```python

# --- Feature Engineering for Future Data ---
# This part is CRITICAL and depends heavily on the features present in X_train
print("Generando características para fechas futuras...")

# 1. Basic Date Features (Replicate from training data prep if they were created there)
# Assuming 'year', 'month', etc., were NOT explicitly added before, extract them now
# If they WERE columns in X_train, ensure they are generated here.
if 'year' in X_train.columns: future_df['year'] = future_df[date_column].dt.year
if 'month' in X_train.columns: future_df['month'] = future_df[date_column].dt.month
if 'day' in X_train.columns: future_df['day'] = future_df[date_column].dt.day # Less likely for monthly
if 'dayofweek' in X_train.columns: future_df['dayofweek'] = future_df[date_column].dt.dayofweek
if 'dayofyear' in X_train.columns: future_df['dayofyear'] = future_df[date_column].dt.dayofyear
if 'weekofyear' in X_train.columns: future_df['weekofyear'] = future_df[date_column].dt.isocalendar().week.astype(int)
if 'quarter' in X_train.columns: future_df['quarter'] = future_df[date_column].dt.quarter

# 2. Lag Features (DEMANDS CAREFUL REPLICATION)
lag_features = [f for f in features if 'lag' in f.lower()]
print(f"Replicando características de lag: {lag_features}")

# Need the full dataset (train + valid) sorted correctly to calculate lags relative to future dates
full_hist_df = pd.concat([train_df, valid_df], ignore_index=True).sort_values(by=[municipality_column, service_type_column, date_column])

# Create a combined future + history frame for easier lag calculation
future_df_with_hist = pd.concat([
    full_hist_df[[date_column, municipality_column, service_type_column, target_column]],
    future_df[[date_column, municipality_column, service_type_column]] # Add future rows without target yet
]).sort_values(by=[municipality_column, service_type_column, date_column]).reset_index(drop=True)

# Calculate lags based on historical TARGET data
for lag_feat in lag_features:
    try:
        # Infer lag amount from feature name (e.g., 'lag_12' -> 12)
        lag_amount = int(lag_feat.split('_')[-1])
        print(f" -> Calculando {lag_feat} (lag {lag_amount})...")
        # Group by series identifiers and shift the target variable
        future_df_with_hist[lag_feat] = future_df_with_hist.groupby([municipality_column, service_type_column])[target_column].shift(lag_amount)
    except ValueError:
        print(f"Advertencia: No se pudo extraer el número de lag de '{lag_feat}'. Omitiendo.")
    except Exception as e:
        print(f"Error calculando lag para '{lag_feat}': {e}. Omitiendo.")


# Filter back to only the future dates, now with calculated lags
future_df_processed = future_df_with_hist[future_df_with_hist[date_column].isin(future_dates)].copy()


# 3. Rolling Features (e.g., rolling means, stddev)
rolling_features = [f for f in features if 'rolling' in f.lower() or 'roll_' in f.lower()]
print(f"Replicando características móviles (rolling): {rolling_features}")
for roll_feat in rolling_features:
     try:
         # Infer window size and aggregation type (e.g., 'rolling_mean_7' -> mean, 7)
         parts = roll_feat.split('_')
         agg_func = parts[-2] # e.g., 'mean', 'std', 'median'
         window_size = int(parts[-1])
         print(f" -> Calculando {roll_feat} (func: {agg_func}, window: {window_size})...")

         # Calculate on the target, shifting by 1 so it only uses past data relative to the prediction date
         grouped = future_df_with_hist.groupby([municipality_column, service_type_column])[target_column]
         # The shift(1) makes sure the rolling window calculation for a given date only uses data *before* that date.
         rolling_calc = grouped.shift(1).rolling(window=window_size, min_periods=max(1, window_size // 2)).agg(agg_func) # Adjust min_periods if needed
         future_df_with_hist[roll_feat] = rolling_calc

     except IndexError:
         print(f"Advertencia: No se pudo extraer la función/ventana de '{roll_feat}'. Omitiendo.")
     except ValueError:
        print(f"Advertencia: No se pudo extraer el número de ventana de '{roll_feat}'. Omitiendo.")
     except Exception as e:
        print(f"Error calculando rolling feature para '{roll_feat}': {e}. Omitiendo.")

# Filter back to future dates again, now potentially with rolling features
future_df_processed = future_df_with_hist[future_df_with_hist[date_column].isin(future_dates)].copy()


# 4. Other Features (Demographics, Economic Indicators, etc.)
# These might be constant per municipality or vary slowly over time.
# Assume they were present in the original X_train and need to be added to future_df.
# Strategy: Use the *last known value* for each municipality/service type combination.
other_features = [f for f in features if f not in lag_features and f not in rolling_features and f not in [municipality_column, service_type_column, 'year', 'month', 'day', 'dayofweek', 'dayofyear', 'weekofyear', 'quarter']]
print(f"Replicando otras características (usando último valor conocido): {other_features}")

if other_features:
    last_known_features = full_hist_df.loc[full_hist_df.groupby([municipality_column, service_type_column])[date_column].idxmax()]
    last_known_features = last_known_features[[municipality_column, service_type_column] + other_features]

    # Merge these last known values into the future dataframe
    future_df_processed = pd.merge(future_df_processed.drop(columns=[f for f in other_features if f in future_df_processed.columns], errors='ignore'),
                                   last_known_features,
                                   on=[municipality_column, service_type_column],
                                   how='left')


# --- Final Check and Prediction ---
# Ensure all required feature columns are present and in the correct order
missing_cols = set(X_train.columns) - set(future_df_processed.columns)
for c in missing_cols:
    print(f"Advertencia: La característica '{c}' falta en los datos futuros después del procesamiento. Añadiendo con valor 0/NaN.")
    future_df_processed[c] = 0 # Or np.nan - model should handle NaNs if trained appropriately

# Fill any NaNs created during lag/rolling calculation (e.g., at the beginning of series)
# Use 0 or a more sophisticated method if needed (e.g., mean/median imputation)
future_df_processed = future_df_processed.fillna(0)

# Reorder columns to match X_train
future_X = future_df_processed[X_train.columns] # Order matters!

# Select the best model/ensemble for prediction
predictor = None
if best_model_name == 'Ensemble':
    print("Usando el ensamble para la predicción final.")
    # Predict with each model in the ensemble and average
    ensemble_final_preds = []
    if 'LGBM' in models: ensemble_final_preds.append(models['LGBM'].predict(future_X))
    if 'XGBoost' in models:
         future_X_xgb = future_X.copy()
         future_X_xgb.columns = ["".join (c if c.isalnum() else "_" for c in str(x)) for x in future_X_xgb.columns]
         ensemble_final_preds.append(models['XGBoost'].predict(future_X_xgb))
    if 'RandomForest' in models: ensemble_final_preds.append(models['RandomForest'].predict(future_X))

    if ensemble_final_preds:
        final_predictions = np.mean(np.array(ensemble_final_preds), axis=0)
    else:
        print("Error: No se pudieron generar predicciones del ensamble.")
        final_predictions = np.zeros(len(future_X)) # Fallback

elif best_model_name in models:
    print(f"Usando el modelo '{best_model_name}' para la predicción final.")
    predictor = models[best_model_name]
    if best_model_name == 'XGBoost':
        future_X_xgb = future_X.copy()
        future_X_xgb.columns = ["".join (c if c.isalnum() else "_" for c in str(x)) for x in future_X_xgb.columns]
        final_predictions = predictor.predict(future_X_xgb)
    else:
        final_predictions = predictor.predict(future_X)
else:
    print("Error: No se encontró el modelo seleccionado para predicción.")
    final_predictions = np.zeros(len(future_X)) # Fallback


# Add predictions to the future dataframe
future_df_processed[target_column] = final_predictions

# Clip predictions at 0 (demand cannot be negative)
future_df_processed[target_column] = np.maximum(0, future_df_processed[target_column]).round(2)

```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from xgboost import XGBRegressor
from scipy.stats import ttest_rel
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings("ignore")

# ===============================
# 1. CARGA Y PREPARACIÓN DE DATOS
# ===============================

train_df = pd.read_csv("healthcare_train_data.csv")
valid_df = pd.read_csv("healthcare_valid_data.csv")

le_mun = LabelEncoder()
le_serv = LabelEncoder()

train_df = train_df[train_df["Municipality"].notna()]
train_df["Municipality_enc"] = le_mun.fit_transform(train_df["Municipality"])
train_df["Service_Type_enc"] = le_serv.fit_transform(train_df["Service_Type"])

valid_df = valid_df[valid_df["Municipality"].isin(train_df["Municipality"].unique())]
valid_df["Municipality_enc"] = le_mun.transform(valid_df["Municipality"])
valid_df["Service_Type_enc"] = le_serv.transform(valid_df["Service_Type"])

# Variables predictoras seleccionadas (refinadas)
features = [
    "Year", "Month", "Municipality_enc", "Service_Type_enc",
    "Service_Count_rolling_mean_12", "Service_Count_rolling_std_12",
    "Growth_Rate_MoM", "Growth_Rate_YoY",
    "Month_Sequence", "Is_Holiday_Month", "Is_Anomaly"
]

target = "Service_Count"

results = []
models = {}

# ==================================
# 2. MODELO POR COMBINACIÓN MUNICIPIO-SERVICIO
# ==================================

for (mun, serv), group in train_df.groupby(["Municipality_enc", "Service_Type_enc"]):
    val_mask = (valid_df["Municipality_enc"] == mun) & (valid_df["Service_Type_enc"] == serv)
    if val_mask.sum() < 6 or group.shape[0] < 12:
        continue

    X_train = group[features]
    y_train = group[target]

    X_valid = valid_df[val_mask][features]
    y_valid = valid_df[val_mask][target]

    model = XGBRegressor(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=4,
        subsample=0.8,
        colsample_bytree=0.8,
        reg_alpha=0.5,
        reg_lambda=0.8,
        early_stopping_rounds=20,
        eval_metric="rmse",
        random_state=42
    )

    eval_set = [(X_train, y_train), (X_valid, y_valid)]
    model.fit(X_train, y_train, eval_set=eval_set, verbose=False)

    y_pred = model.predict(X_valid)

    mae = mean_absolute_error(y_valid, y_pred)
    rmse = np.sqrt(mean_squared_error(y_valid, y_pred))
    r2 = r2_score(y_valid, y_pred)

    # Prueba Diebold-Mariano (versión simplificada)
    y_naive = y_valid.shift(1).fillna(method='bfill')
    naive_error = (y_valid - y_naive) ** 2
    model_error = (y_valid - y_pred) ** 2
    t_stat, p_val = ttest_rel(naive_error, model_error)

    results.append({
        "Municipality": le_mun.inverse_transform([mun])[0],
        "Service_Type": le_serv.inverse_transform([serv])[0],
        "MAE": mae,
        "RMSE": rmse,
        "R2": r2,
        "DieboldMariano_p": p_val,
        "Count": len(y_valid)
    })

    models[(mun, serv)] = model

    # ==================================
    # 3. VISUALIZACIÓN (solo para 1 ejemplo)
    # ==================================
    if le_mun.inverse_transform([mun])[0] == "VILLAVICENCIO" and le_serv.inverse_transform([serv])[0] == "HOSPITALARIA":
        # Importancia de variables
        importances = model.feature_importances_
        sorted_idx = np.argsort(importances)

        plt.figure(figsize=(8, 5))
        plt.barh([features[i] for i in sorted_idx], importances[sorted_idx])
        plt.title("Importancia de variables")
        plt.xlabel("Gain")
        plt.tight_layout()
        plt.show()

        # Curva de aprendizaje
        evals = model.evals_result()
        plt.plot(evals["validation_0"]["rmse"], label="Train RMSE")
        if "validation_1" in evals:
            plt.plot(evals["validation_1"]["rmse"], label="Valid RMSE")
        plt.title("Curva de Aprendizaje")
        plt.xlabel("Iteraciones")
        plt.ylabel("RMSE")
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()

# ==================================
# 4. RESULTADOS FINALES
# ==================================

results_df = pd.DataFrame(results)
filtered_results = results_df[results_df["Count"] >= 12].sort_values(by="RMSE")
print(filtered_results.head(10))
```
